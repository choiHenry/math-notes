# Proof of Theorem 2.1 in Rust(1985a)

*  A durable can be represented quite generally as a discrete parameter Markov process whose state
at time $t$ is represented by a nonnegative real number $x_t$.
*  A new durable consists of the triple $\{P, m(. ; \tau), \Phi(\cdot, \cdot)\}$ where $\tau$
is a parameter indexing a particular consumer, $P$ is the sales price of a new durable set by the producer, $m(x; r)$ is the conditional expectation of one period operating and maintenance costs for consumer $\tau$ as a function of the asset's state $x$, and $\Phi(x, y)$ is the conditional probability distribution of next period's condition $y$ as a function of current condition $x$.
* There are a continuum of consumers distinguished by a characteristic $\tau$ where $\tau$ is a real number in the interval [$\underline\tau$, $\overline\tau$l. The population distribut of consumers is given by a probability distribution $H$ on the interval [$\underline\tau$, $\overline\tau$l, wher e $H(\tau)$ is the fraction of the population whose preference parameter is less than $\tau$.
* We assume that the per period utility of having an asset in condition $x$ and income $I$ is given by $U(I, x; \tau)=a(\tau)I+q(x;\tau)$.
* $m(x; \tau)$: conditional expectation of 1 period maintenance costs for asset $x$;
* $P$: purchase price of an asset in condition $z$.
* Formally, consumer $\tau$ who initially owns a durable in condition $x_0 = x$ seeks an infinite sequence of decision-rules $u$, which attain the infimum $j_\tau(x)$ given by
$$\tag{2.3} \sup_\Pi E\{\Sigma_{t=0}^\infty\beta^tg_r(x_t, u_t)|x\}$$
where
$$0<\beta < 1,$$
$$\Pi=(u_0, u_1, \dots),$$
and where:
$$g_\tau(x_t, u_t)=\begin{cases}a(\tau)[I_t-m(x_t;\tau) + q(x_t;\tau), \text{ if } u_t(x_t)=x_t,\\a(\tau)[I_t-P(u_t)+T(x_t)-m(u_t;\tau)]+q(u_t;\tau), \text{ otherwise }.\end{cases}$$
* Define an "operating cost" function $M(\cdot; \tau)$ by $M(x; \tau) = m(x; \tau) - q(x; \tau)/a(\tau)$.
* Then a solution $\Pi$ to (2.3) will exist iff $\Pi$ is a solution to
$$\tag{2.4} J_\tau(x)=\inf_\Pi E\{\Sigma_{t=0}^\infty\beta^tG_\tau(x_t, u_t)|x\}$$,
where
$$G_\tau(x_t, u_t)=\begin{cases}M(x_t;\tau)\text{ if } u_t(x_t)=x_t,\\P(u_t)-P(x_t)+M(u_t;\tau) \text{ otherwise }.\end{cases}$$
* Note that the transition probability $\Phi$ defines a continuous linear transformation $E$ on the Banach space of bounded Borel measurable functions defined by $Ef(x) = \int_0^\infty f(y)\phi(x, dy)$. $E$ is the conditional expectation operator generated by $\Phi$.
* To solve the consumer's problem (2.4) we must solve for the value function $J_\tau$ which is the unique solution to Bellman's equation given by
$$J_\tau = \min[M(x;\tau)+\beta EJ_\tau(x), \inf_{0\leq z\leq \gamma^*}\{P(z)-P(x)+M(z;\tau)+\beta EJ_\tau(z)\}\tag{2.6}$$
* (Assumption A1): *There are no taxes or transaction costs and all trading occurs at equilibrium prices; there is no individual bargaining over price*.
* (Assumption A2): *Each consumer can costlessly verify the physical condition of each asset, and knows the structure of secondary market prices $P$, and the technological characteristics of durables, $\Phi$*.
* (Assumption A3): *For each $x\in \mathbb{R}^+$, $\Phi(x, \cdot)$ is an absolutely continuous distribution function which satisfies $\Phi(x, 0)=0$*.
* (Assumption A4): *For each $y\in \mathbb{R}^+$, $\Phi(\cdot, y)$ is a continuous, nondecreasing function.*

## Theorem 2.1
*Suppose that functions $P$ and $-M(\cdot, \tau)$ are nonincreasing and continuous with $P(x)=\underline{P}$ for $x\geq \gamma^*$. Then under Assumptions A1-A4*,
$$\begin{equation}\tag{2.7}u(x_t, \tau)=z^*(\tau)\end{equation}$$
where $z^*(\tau)$ is any element of the set $Z^*(\tau)$ defined by
$$Z^*(\tau)=argmin_{0\leq z \leq gamma^*}M(z; \tau)+ P(z) - \beta EP(z)\tag{2.8}$$
and where $J_\tau$ is the unique, bounded continuous solution to (2.6) given by
$$J_\tau = [1/(1-\beta)][P(z^*(\tau))+M(z^*(\tau);\tau)-\beta EP(z^*(\tau))]-P(x)\tag{2.9}$$

## Proof

Since the set of available assets is the compact interval $[0, \gamma^*]$, there exists an optimal replacement asset $z^*(T)$, i.e., an asset which attains the infimum in (2.6). Suppose for $x \neq z^*(T)$ it is optimal to keep $x$; then (2.6) implies
$$P(z^*(\tau))+M(z^*(\tau);\tau)+\beta EJ_\tau(z^*(\tau))>P(x)+M(x;\tau)+\beta EJ_\tau(x).\tag{1}.$$
This contradicts the definition of $z^*(\tau)$ since $z^*(\tau)$ satisfies for any $z$
$$P(z^*(\tau))-P(x)+M(z^*(\tau);\tau)+\beta EJ_\tau(z^*(\tau))\leq P(z)-P(x)+M(z;\tau)+\beta EJ_\tau(z).\tag{2}$$ Thus $u_t(x)=z^*(x)$ for all $x$. Using this fact (2.6) reduces to
$$J_\tau(x)=P(z^*(\tau))-P(x)+M(z^*(\tau);\tau)+\beta EJ_\tau(z^*(\tau)).\tag{3}$$
Taking expectations on both sides of (3) and solving for $EJ_\tau(z^*(\tau))$ yields the solution for $J_\tau$ given in (2.9).
First taking expectations on both sides of (3) yields
$$EJ_\tau(x) = P(z^*(\tau))-EP(x)+M(z^*(\tau);\tau)+\beta EJ_\tau(z^*(\tau)).\tag{4}$$
Subsituting $z^*(\tau)$ to $x$ in (4),
$$EJ_\tau(z^*(\tau)) = P(z^*(\tau))-EP(z^*(\tau))+M(z^*(\tau);\tau)+\beta EJ_\tau(z^*(\tau)).\tag{5}$$
Solving (5) for $EJ_\tau(z^*(\tau))$,
$$EJ_\tau(z^*(\tau)) = [1/(1-\beta)][P(z^*(\tau))-EP(z^*(\tau))+M(z^*(\tau);\tau).\tag{6}$$
Subsituting (6) into (2) yields (2.9).
Taking expectations on (2.9),
$$EJ_\tau(x)=[1/(1-\beta)](P(z^*(\tau))+M(z^*(\tau);\tau)-\beta EP(z^*(\tau)))-EP(x).\tag{7}$$
Substituting $EP\tau(z)$ in $\inf$ of (2.6) yields (2.8).
